```python
#Work on pre train classification CNN based classification
from tensorflow.keras.applications import vgg16
model = vgg16.VGG16() # Load the pretain CNN based classifier
model.summary(show_trainable = True)#desplay the architecture of the load model
```

```python
base_model = vgg16.VGG16( input_shape=(32,32,3), weights= 'imagenet', include_top = False)
base_model.summary()
```

```python
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.models import Model
inputs = base_model.inputs
x = base_model.output
x = Flatten()(x)
x = Dense(16, activation = 'relu')(x)
outputs = Dense(2, activation = 'softmax')(x)
binary_model = Model(inputs = inputs, outputs = outputs)
binary_model.summary(show_trainable = True)
```

```python
for layer in binary_model.layers[:-2]:
    layer.trainable = False
binary_model.summary(show_trainable=True)
```
